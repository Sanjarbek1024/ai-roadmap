Principal Component Analysis

```python
from sklearn.datasets import fetch_openml
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
```

```python
mnist = fetch_openml('mnist_784')
```

```python
X = mnist.data.values
y = mnist.target.values
target_names = mnist.target_names
```

```python
import numpy as np

target_names = np.unique(y)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X_reduced = pca.fit_transform(X)
```

```python
fig = plt.figure(figsize=(16, 12))
ax = fig.add_subplot(111, projection='3d')

colors = [
    'red',
    'green',
    'blue',
    'orange',
    'purple',
    'brown',
    'pink',
    'gray',
    'cyan',
    'magenta'
]
for target, color, label in zip(target_names, colors, target_names.astype(str)):
    ax.scatter(X_reduced[y == target, 0], X_reduced[y == target, 1], X_reduced[y == target, 2], c=color, label=label)

ax.legend()
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=3)
X_reduced_tsne = tsne.fit_transform(X)
```

```python
fig = plt.figure(figsize=(16, 12))
ax = fig.add_subplot(111, projection='3d')

colors = [
    'red',
    'green',
    'blue',
    'orange',
    'purple',
    'brown',
    'pink',
    'gray',
    'cyan',
    'magenta'
]
for target, color, lab l in zip(target_names, colors, target_names.astype(str)):
    ax.scatter(X_reduced_tsne[y == target, 0], X_reduced_tsne[y == target, 1], X_reduced_tsne[y == target, 2], c=color, label=label)

ax.legend()
plt.show()
```
